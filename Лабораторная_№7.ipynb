```python
import numpy as np
import matplotlib.pyplot as plt
#Создание функции, считающей производную
def diff(x,y):
  N=len(x)  #Число точек в массиве x (общее число точек исходной функции)
  res=np.zeros(N)  #Создаем массив под значения производной (длина N)
  #Последнее число в массиве будет 0
  for k in range(N-1):
    res[k]=(y[k+1]-y[k])/(x[k+1]-x[k])
  return res
#Создание дискретной тестовой функции
fs=1000 #Частота дискретизации
h=1/fs  #Шаг по оси t
x=np.arange(0,10+h,h)
y=np.exp(-2*x)*np.sin(x*314*0.5)
#тестирование программы на квадратичной функции
#y=x**2
#тестирование программы на линейной функции
#y=x+1
#создание цикла для расчета интеграла функции
s=0
S=0
#вызываем функцию с помощью оператора def для подсчета интеграла
def integral(x,y):
  global S
  N=len(x)
  res=np.zeros(N)
  for i in range(N-1):
    res[i]=y[i]*(x[i+1]-x[i])
    S=S+res[i]
  return res
#Строим график исходной функции
plt.figure(1)
plt.plot(x,y)
plt.grid()
plt.title('График исходной функции')
#Вызываем функцию для расчета производной
dy=diff(x,y)
#Строим график производной 
plt.figure(3)
plt.plot(x,dy)
plt.grid()
plt.title('График производной')
#вызовем функцию для расчета интегральной функции
dxy=integral(x,y)
#выведем значение интеграла функции
print(S)
```
